```python
# Set parameters
img_height = 200
img_width = 200
# Build datasets to train
## Create preprocessing funtion
def preprocessing(file_path):
  ### read file
  img = tf.io.read_file(file_path)
  img = tf.image.decode_jpeg(img, channels= 3)
  ### Transform: Normalize img having the interval of [0,1]
  img = tf.image.convert_image_dtype(img, tf.float32)
  ### Resize
  img = tf.image.resize(img, [img_width, img_height])
  label = get_label(file_path)
  return img, label

ds_size = 20856
batch_size = 20
#Build datasets
data = file_list_ds.shuffle(ds_size)
data = data.map(preprocessing, num_parallel_calls=tf.data.AUTOTUNE)
data = data.take(ds_size)
data = data.prefetch(1)
# Divide teh data into test and train
TRAIN_TEST_SPLIT = 0.2
num_examples_in_test_ds = int(data.cardinality().numpy() * TRAIN_TEST_SPLIT)

data = data.shuffle(ds_size)
test_ds = data.take( num_examples_in_test_ds )
train_ds = data.skip( num_examples_in_test_ds )

test_ds=test_ds.batch(batch_size)
train_ds=train_ds.batch(batch_size)
```

```python
from tensorflow.keras import layers
from keras.models import load_model
import os
# Create Functional Model
model_input = tf.keras.Input(shape=(200,200,3))
x = layers.Conv2D(96, 3, strides = (1,1), padding = 'valid', activation = 'relu')(model_input)
x = layers.MaxPool2D(pool_size = (3,3), strides = 2)(x)
x = layers.Conv2D(128, 3, strides = (1,1), padding = 'valid', activation = 'relu')(x)
x = layers.MaxPool2D(pool_size = (3,3), strides = 2)(x) 
x = layers.Conv2D(256, 3, strides = (1,1), padding = 'valid', activation = 'relu')(x)
x = layers.MaxPool2D(pool_size = (3,3), strides = 2)(x)
x = layers.Conv2D(384, 3, strides = (1,1), padding = 'valid', activation = 'relu')(x)
x = layers.MaxPool2D(pool_size = (3,3), strides = 2)(x)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(512, activation='relu')(x)
model_output = layers.Dense(23, activation='softmax')(x)
model = tf.keras.Model(inputs=model_input, outputs=model_output )
# # tf.keras.utils.plot_model(model, 'model.png', show_shapes=True)
model.load_weights('/content/gdrive/MyDrive/model68epochs.h5')
# print(os.getcwd())
```

```python
#Complie Model
model.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ["accuracy"]
)
```

```python
from keras.callbacks import ModelCheckpoint

filepath="weights-improvement-{epoch:02d}-{accuracy:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, 
                             monitor = 'val_accuracy', 
                             verbose = 1, 
                             save_best_only = True, 
                             mode='auto'
                             )
callbacks_list = [checkpoint]
#Training
history = model.fit(train_ds, epochs=2, validation_data=test_ds,callbacks = callbacks_list)
```
